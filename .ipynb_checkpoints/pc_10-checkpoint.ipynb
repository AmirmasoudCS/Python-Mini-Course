<h2>Reading a File ("r")</h2>

```python
filename = "students.txt"
file = open(filename, "r")
```

<h3>Using <i>with</i> </h3>

```python
with open("students.txt", "r") as f:
    content = f.read()
```

```python
content
```

```python
with open("students.txt", "r") as file:
    lines = file.readlines()
```

```python
lines
```
