<h2>Multiple Assignments</h2>

```python
a = 5
b = 10
a + b
```

<h2>Swap</h2>

```python
a = 2
b = 7
a, b
```

<h3>Method 1</h3>

```python
hold = a
a = b
b = hold
a, b
```

<h3>Method 2</h3>

```python
x, y = 2, 7
x, y
```

```python
x, y = y, x
x, y
```
