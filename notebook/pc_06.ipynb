<p>A loop is a block of code that runs again and again until a condition stops it.</p>
<ol>
    <li> for loop → iterate over a sequence</li>
    <li> while loop → run while the condition is True</li>
</ol>

<h2>The <i>for</i> loop</h2>

```python
for i in range(0,6): 
    print(i)
```

```python
fruits = ["Apple", "Banana", "Mango"]
for fruit in fruits:
    print(fruit)
```

```python
for i in range(10):
    print(i, end=",")
```

<h2>The <i>while</i> loop</h2>

```python
i = 0
while i < 5:
    print(i)
    i += 1
```
